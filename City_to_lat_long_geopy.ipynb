```
from google.colab import files 
import pandas as pd 
import io 
import numpy as np
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

uploaded = files.upload()
```

```
dataset = pd.read_csv(io.BytesIO(uploaded['cities to be cleaned with geopy.csv'])) 
print(dataset.info())
```

```
def do_geocode(address):
    geopy = Nominatim(user_agent="geoapi_eShipz")
    try:
        return geopy.geocode(address,exactly_one=True)
    except GeocoderTimedOut:
        return do_geocode(address)

#Creating Geocoded Location column
dataset['GeocodedLocation']=dataset['source_city'].apply(lambda x: do_geocode(x) if x != None else None)

#Create the Latitude Column
lat=[]
for i in dataset['GeocodedLocation']:
    if i== None:
        lat.append(None)
    else:
        lat.append(i.latitude)
dataset['Source_Latitude']=lat
dataset['Source_Latitude'].astype('float')

#Create the Longitude Column
long=[]
for i in dataset['GeocodedLocation']:
    if i== None:
        long.append(None)
    else:
        long.append(i.longitude)
dataset['Source_Longitude']=long
dataset['Source_Longitude'].astype('float')

#Drop GeocodedLocation Column
dataset=dataset.drop(['GeocodedLocation'],axis=1)

dataset.sample(5)
```
